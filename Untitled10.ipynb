```python
#Create an interactive line chart to show sales trends over the years
import matplotlib.pyplot as plt
from ipywidgets import interact, Dropdown

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year

sales_by_year = df.groupby('Year')['Sales'].sum().reset_index()

def plot_sales_trend(highlight_year=None):
    plt.figure(figsize=(10,6))
    
    plt.plot(sales_by_year['Year'], sales_by_year['Sales'], marker='o', linestyle='-')
    
    colors = ['red' if year == highlight_year else 'blue' for year in sales_by_year['Year']]
    plt.scatter(sales_by_year['Year'], sales_by_year['Sales'], color=colors, s=80)
    
    for i, row in sales_by_year.iterrows():
        plt.text(row['Year'], row['Sales'] + sales_by_year['Sales'].max()*0.01,
                 f"${row['Sales']:,.0f}", ha='center', va='bottom',
                 fontsize=9, color='red' if row['Year']==highlight_year else 'black')
    
    plt.title("Sales Trend Over the Years")
    plt.xlabel("Year")
    plt.ylabel("Total Sales ($)")
    plt.xticks(sales_by_year['Year'])
    plt.grid(True, linestyle='--', alpha=0.5)

    max_sales = sales_by_year['Sales'].max()
    plt.ylim(0, max_sales * 1.15)
    plt.tight_layout()

    plt.show()

interact(plot_sales_trend,
         highlight_year=Dropdown(
             options=[None] + sales_by_year['Year'].tolist(),
             value=None,
             description='Highlight Year:'
         ))

```

```python
#Build an interactive map to visualize sales distribution by country.

df.columns = df.columns.str.strip()

sales_by_country = df.groupby('Country')['Sales'].sum().reset_index()

country_to_iso = {
    'United States': 'USA',
    'Mexico': 'MEX',
    'Canada': 'CAN',
    'France': 'FRA',
    'Germany': 'DEU',
    'United Kingdom': 'GBR',
    'Australia': 'AUS',
    'China': 'CHN',
    'India': 'IND',
}

sales_by_country['ISO-3'] = sales_by_country['Country'].map(country_to_iso)


fig = px.choropleth(
    sales_by_country.dropna(subset=['ISO-3']), 
    locations="ISO-3",
    locationmode="ISO-3",         
    color="Sales",
    hover_name="Country",
    projection="natural earth", 
    title="Interactive Sales Distribution by Country",
    color_continuous_scale="Plasma"
)

fig.show()
```

```python
#Use Seaborn to generate a bar chart showing top 10 products by sales.

df.columns = df.columns.str.strip()

try:
    sales_by_product = df.groupby('Product Name')['Sales'].sum().reset_index()
    top_10_products = sales_by_product.sort_values(by='Sales', ascending=False).head(10)
except KeyError as e:
    print(f"Ошибка: Не удалось найти столбец {e}. Проверьте точное название столбца в вашем файле (например, 'Product Name' vs 'Product_Name').")
    exit()

plt.figure(figsize=(12, 6))


sns.barplot(
    x='Sales', 
    y='Product Name', 
    data=top_10_products,
    hue='Product Name', 
    legend=False,       
    palette='viridis'   
)

plt.title('Top 10 Products by Total Sales (US Superstore)', fontsize=16)
plt.xlabel('Total Sales ($)', fontsize=12)
plt.ylabel('Product Name', fontsize=12)

plt.tight_layout()

plt.show()
```

```python
#Create a scatter plot to analyze the relationship between profit and discount.

df.columns = df.columns.str.strip()

REQUIRED_COLUMNS = ['Profit', 'Discount']
for col in REQUIRED_COLUMNS:
    if col not in df.columns:
        print(f"Ошибка: Не удалось найти столбец '{col}'. Проверьте точное название столбца в вашем файле.")
        exit()

plt.figure(figsize=(10, 6))

sns.scatterplot(
    x='Discount', 
    y='Profit', 
    data=df,
    alpha=0.6,         
    s=15,               
    color='#0077B6'     
)

plt.title('Relationship Between Profit and Discount (Superstore Data)', fontsize=16)
plt.xlabel('Discount Percentage (0.0 to 1.0)', fontsize=12)
plt.ylabel('Profit ($)', fontsize=12)

plt.axhline(0, color='grey', linestyle='--', linewidth=1)

plt.tight_layout()

plt.show()
```

```python
#Compare the insights gained from Matplotlib and Seaborn visualizations.

Matplotlib and Seaborn provide different types of insight even when visualizing the same data, such as the relationship between Profit and Discount.

Matplotlib focuses on showing the raw data as clearly as possible. It allows you to see the exact distribution of points, making it easier to detect outliers, extreme losses, and irregular patterns. When using Matplotlib, you can clearly observe that higher discounts often lead to negative profits and that profit values vary significantly even at similar discount levels. This makes it ideal for deep analytical work where precision and full control are important.

Seaborn, on the other hand, emphasizes interpretation and visual clarity. It automatically enhances the chart by adding smoothing and trend lines, which makes overall patterns more obvious. With Seaborn, the negative relationship between discount and profit becomes immediately clear, helping you understand the general business impact of discounting strategies without focusing too much on individual anomalies. This makes Seaborn better suited for presentations and high-level insights.

In summary, Matplotlib is better for detailed exploration and identifying specific data issues, while Seaborn is better for communicating overall trends and telling a clear visual story. Using both together provides the most complete understanding of the data.
```

```python
#Document your observations about the ease of use and effectiveness of both tools.

Matplotlib requires more manual effort to produce visually appealing charts. The user must explicitly control elements such as colors, labels, grid styles, and layouts. While this can feel complex for beginners, it offers very precise control and flexibility. This makes Matplotlib highly effective for detailed analysis, custom visual design, and situations where exact formatting is required. However, the learning curve is steeper, and creating polished visuals often takes more time and code.

Seaborn is noticeably easier to use, especially for statistical visualizations. It provides attractive default styles and automatically handles many visual enhancements such as smoothing, grouping, and trend lines. This reduces the amount of code needed and speeds up the visualization process. As a result, Seaborn is more effective for quickly identifying patterns, trends, and relationships in data, particularly when the goal is clear communication rather than granular control.

Overall, Matplotlib excels in flexibility and precision but demands more technical effort, while Seaborn prioritizes simplicity and readability, making it more efficient for rapid analysis and presentation-ready visuals. The most effective workflow combines both: Seaborn for fast insight and Matplotlib for fine-tuned customization.
```
